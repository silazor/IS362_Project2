### Show Firearm permits by state. Which states have the highest, lowest firearm permits issued?

Kentucky, Louisiana, Maine, Mariana Islands, Maryland, Massachussets, Mississippi, Alaska, Montana, and Missouri are all tied for the top 10 states having the least amount of firearm permits issued.

North Carolina has issued the most firearm permits.

```python
import pandas
import os

os.chdir('c:/Users/silaz/OneDrive/Documents/IS362/w6')

nics_firearm = pandas.read_csv('./nics-firearm-background-checks.csv')
nics_grouped_by_state_month=(nics_firearm.groupby(['state', 'month'])['permit'].sum())
nics_grouped_by_state_month.reset_index().sort_values(['permit', 'state'], ascending = False)

print(nics_firearm.groupby(['state']).permit.agg(['min']).sort_values(by = 'min', ascending=False).tail(10))
print(nics_firearm.groupby(['state']).permit.agg(['max']).sort_values(by = 'max', ascending=False).head(10))
```

### What country suspected of a Zika virus outbreak has the most laboratory confirmed cases?

Columbia had the most suspected cases of the Zika virus with 82,188 cases.

```python
paho_reported = pandas.read_csv("./paho-who-cases-reported-2016-03-02.csv")
paho_reported.columns = paho_reported.columns.str.strip().str.lower().str.replace(' ', '_')
paho_reported['suspected_cases'] = paho_reported['suspected_cases'].str.replace(',', '')
paho_reported['suspected_cases'] = pandas.to_numeric(paho_reported['suspected_cases'])
paho_reported.groupby(['country_/_territory'])['suspected_cases'].sum().sort_values(ascending=False).head(10)
```
